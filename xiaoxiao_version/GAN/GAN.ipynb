```python
from torchvision import datasets, transforms
import torch
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import numpy as np
import pandas as pd
import torch.utils.data as Data
from torch.utils.data.dataloader import DataLoader
import torch.nn as nn

ART_COMPONENTS = 10
BATCH_SIZE = 64
PAINT_POINTS =  np.array([list(np.linspace(-1,1,ART_COMPONENTS)) for _ in range(BATCH_SIZE)])
N_IDEALS = 5

def artist_works():     # painting from the famous artist (real target)
    a = np.random.uniform(1, 2, (BATCH_SIZE,1))
    paintings = a * np.power(PAINT_POINTS, 2) + (a-1)
    paintings = torch.tensor(paintings)
    return paintings

G = nn.Sequential(nn.Linear(N_IDEALS, 128),
                  nn.ReLU(),
                  nn.Linear(128, ART_COMPONENTS)
                 )
D = nn.Sequential(nn.Linear(ART_COMPONENTS, 128),
                 nn.ReLU(),
                 nn.Linear(128,1),
                 nn.Sigmoid()
                 )
LR_D = 0.0001
LR_G = 0.0001

opt_D = torch.optim.Adam(D.parameters(), lr=LR_D)
opt_G = torch.optim.Adam(G.parameters(), lr=LR_G)


for step in range(1000):
    artist_paintings = artist_works()
    G_ideas = torch.randn(BATCH_SIZE, N_IDEALS) # 生成N个batch & 每个batch生成N_IDEALS个点
    G_gen_paintings = G(G_ideas)
    D_generate = D(G_gen_paintings)
    D_artist = D(artist_paintings.float())
    
    D_loss = - torch.mean(torch.log(D_artist)+torch.log(1-D_generate))
    G_loss = torch.mean(torch.log(1. - D_generate))
    
    opt_D.zero_grad()
    D_loss.backward(retain_graph = True)
    opt_D.step()
    
    opt_G.zero_grad()
    G_loss.backward(retain_graph = True)
    opt_G.step()
    
    print(G_loss)
```
